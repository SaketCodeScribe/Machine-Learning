```python
#to replace values in a column
d={'no':False, 'yes':True}
df['international plan']=df['international plan'].map(d)
df.head()
```

```python
#Same thing can be done with the replace method
df=df.replace({'voice mail plan':d})
```

```python
df.head()
```

```python
#grouping data in Pandas
#First, the groupby method divides the grouping_columns by their values. They become a new index in the resulting dataframe.
#Then, columns of interest are selected (columns_to_show). If columns_to_show is not included, 
#all non groupby clauses will be included.
#Finally, one or several functions are applied to the obtained groups per selected columns.

columns_to_show=['total day minutes','total eve minutes', 'total night minutes']
df.groupby(['churn'])[columns_to_show].describe(percentiles=[])
```

```python
columns_to_show=['total day minutes','total eve minutes', 'total night minutes']
df.groupby(['churn'])[columns_to_show].agg([np.mean,np.std,np.min,np.max])
```

```python
#summary table
pd.crosstab(df['churn'],df['international plan'])
```

```python
pd.crosstab(df['churn'],df['voice mail plan'],normalize=True)
```

```python
#average numbers of day, evening and night calls by area code
df.pivot_table(['total day calls','total eve calls','total night calls'],['area code'],aggfunc='mean')
```

```python
total_calls = df['total day calls'] + df['total eve calls'] + df['total night calls'] + df['total intl calls'] 
df.insert(loc=len(df.columns),column='total calls',value='total calls')
df.head()
```

```python
df['total charge']=df['total day charge'] + df['total eve charge']+df['total night charge'] + df['total intl charge']
df.head()
```

```python
#o delete columns or rows, use the drop method, passing the required indexes 
#and the axis parameter (1 if you delete columns, and nothing or 0 if you delete rows). 
#The inplace argument tells whether to change the original DataFrame. With inplace=False, 
#the drop method doesn't change the existing DataFrame and returns a new one with 
#dropped rows or columns. With inplace=True, it alters the DataFrame.

# get rid of just created columns 
df.drop(['total charge', 'total calls'], axis=1, inplace=True) 
```

```python
df['many service calls']=(df['customer service calls']>3).astype(int)
pd.crosstab(df['many service calls'],df['churn'],margins=True)
```

```python
pd.crosstab(df['many service calls']&df['international plan'],df['churn'])
```

```python
df['many service calls & international plan']= (df['many service calls']&df['international plan'])
sns.countplot(x='many service calls & international plan',hue='churn',data=df);
```

```python
pd.crosstab(df['many service calls & international plan'],df['churn'],margins=True)
#Therefore, predicting that a customer will churn (Churn=1) in the case when the number 
#of calls to the service center is greater than 3 and the International Plan is added 
#(and predicting Churn=0 otherwise), we might expect an accuracy of 85.8% (we are mistaken 
#only 464 + 9 times). This number, 85.8%, that we got with very simple reasoning serves 
#as a good starting point (baseline) for the further machine learning models that we will build.
```
